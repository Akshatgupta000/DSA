```python
def recurfact(n):
    if n==1 or n==0:
        return 1
    else:
        return n*recurfact(n-1)
a=int(input("a: "))
if a<0:
    print("factorial not exist for -ve")
else:
    print("factorial:",recurfact(a) )
    
```

```python
def F(n):
    if n==0:
        return 1
    if n==1:
        return 0
    return F(n-2)
Z=int(input("a: "))
a=F(Z)
if(a==1):
    print("Even")
else:
    print("odd")
    
```
