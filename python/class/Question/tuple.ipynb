```python
# write a program to accept values from user in a tuple. Add a tuple to it and display its element one by one. Also display its maximum and minimum value.

tup=tuple()
n=int(input("Enter any no. "))
for i in range(n):
    a=int(input("Enter no."))
    tup=tup+(a,)
print(tup)
print("maximum value- ",max(tup))
print("minimum value- ",min(tup))
```
