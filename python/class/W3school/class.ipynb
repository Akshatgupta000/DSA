```python
# convert into uppercase

class convert:
    def __init__(self,string):
        self.string=string
    def uppercase(self):
        return self.string.upper()
        

string=input("Enter the name: ")
a=convert(string)
print(a.uppercase())

        
```

```python
# reverse the string

class ReverseString:
  def __init__(self, string):
    self.string = string

  def reverse(self):
    return self.string[::-1]

string = input()
a = ReverseString(string)
print(a.reverse())
```

```python
class measure:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def area(self):
        return self.length*self.breadth

length=int(input())
breadth=int(input())
a=measure(length,breadth)
b=a.area()
print(b)
```
