```python
# program to check whether a number is prime or not
n=int(input("Enter a number"))
k=0
if(n==0 or n==1):
    print("Not a prime number")
else:
    i=2
    while(i<n):
        k=k+1
        i=i+1
    if(k==0):
        print(n,"is prime")
    else:
        print(n,"not prime")
        
```

```python
# fibonacci
n=int(input("Enter: "))
n1,n2=0,1
count=0

while(count<n):
    print(n1)
    nth=n1+n2
    n1=n2
    n2=nth
    count=count+1
```
