```python
class vehicle:
    def setenginemodel(self,engine):
        self.engine=engine
    def getenginemodel(self):
        print(self.engine)
    
class Honda(vehicle):
    def setvehiclemodel(self,model):
        self.model=model 
    def getvehiclemodel(self):
        print(self.model)
myvehicle=Honda()
myvehicle.setenginemodel("EK_1")
myvehicle.setvehiclemodel("V9")
print("vehicle details:")
myvehicle.getenginemodel()
myvehicle.getvehiclemodel()

    
```
