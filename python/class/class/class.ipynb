```python
class hi():
    roll=0
    mark1=0
    mark2=0
    mark3=0
    def read(self):
        self.roll=int(input("Enter your roll no.: "))
        self.mark1=int(input("Enter mark1:"))
        self.mark2=int(input("Enter mark2:"))
        self.mark3=int(input("Enter mark3:"))
    def display(self):
        print("Roll no.",self.roll)
        print("avg is",(self.mark1+self.mark2+self.mark3)/3)
b=hi()
b.read()
b.display()
```

```python
class employee():
    def __init__(self,name,age,id):
        self.name=name
        self.age=age
        self.id=id
        
    def display(self):
        print("Name is: ",self.name)
        print("age is: ",self.age)
        print("id is: ",self.id)
name=input("Enter your name")
age=int(input("Enter your age"))
id=int(input("Enter your id"))

obj=employee(name,age,id)
obj.display()
        
```
