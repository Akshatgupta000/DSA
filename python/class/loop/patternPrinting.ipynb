```python
print("Multiplication Table")
print(" |",end="")
for j in range(1,11):
    print(" ",j,end="")
print()
print("________________________________________________________________________________")

for i in range(1,11):
    print(i,"|",end="")
    for j in range(1,11):
        print(format(i * j,"4d"),end="")
    print()
```
