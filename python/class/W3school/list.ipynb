```python
# Python: Sum all the items in a list

def sum_list(item):
    sum=0
    for x in item:
        sum+=x
    return sum
print(sum_list([1,2,3,4]))
```

```python
# Python: Multiply all the items in a list  

def mul_list(item):
    mul_numbers=1
    for x in item:
        mul_numbers*=x
    return mul_numbers
print(mul_list([1,2,3,4]))
```
