```python
num1=10
num2=2

# Arithmetic Operators
print("num1 + num2 is", num1 + num2)
print("num1 - num2 is", num1 - num2)
print("num1 * num2 is", num1 * num2)
print("num1 / num2 is", num1 / num2)
print("num1 % num2 is", num1 % num2)     # remainder.
print("num1 // num2 is", num1 // num2)   #point ke baad value nhi dikhegi.
print("num1 ** num2 is", num1 ** num2)   #power function 10 to the power 2 cal.
```

```python
#assignment Operators

a = 4
a /= 2
# a+=2
print(a)
```

```python
# Comparison Operators

x = 8
y = 3
z = 8
print(x>y)
print(x<y)
print(x!=y)
print(x==z)
```

```python
x = 8
y = 3
z = 8

# logical operators
print(x==z and x<y)
print(x==z or x<y)
print(not(False))
print(not(True))
```
