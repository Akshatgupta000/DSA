```python
# faactorial
def factorial(n):
    if n==1 or n==0:
        return 1
    else:
        return n*factorial(n-1)
a=int(input())
if(a<0):
    print("not exist for negative numbers")
else:
    print(factorial(a))
```

```python
a=int(input())
fact=1
i=1
while(i<=a):
    fact=fact*i
    i+=1
print(fact)
```

```python
def reverse(n,x):
    if(n==0):
        return x
    x=(x*10)+(n%10)
    return reverse(n//10,x)
num=int(input("x: "))
print(reverse(num,0))
```
