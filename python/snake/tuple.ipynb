```python
# change tuple values  ->convert the tuple into a list, change the list, and convert the list back into a tuple.
x=("apple","banana","mango")
y=list(x)
y[1]="kiwi"
x=tuple(y)
print(x)
```

```python
# add items
x=("apple","banana","mango")
y=list(x)
y.append("kiwi")
x=tuple(y)
print(x)
```

```python
# we can add two tuples
a=("apple","banana","mango")
b=("kiwi",)
a+=b
print(a)
```

```python
a=("apple","banana","mango")
b=list(a)
b.remove("apple")
a=tuple(b)
print(a)
```

```python
# loop through tuple
a=("apple","banana","mango")
for x in a:
    print(x)
```

```python
# loop through the index
a=("apple","banana","mango")
for i in range(len(a)):
    print(a[i])
```

```python
# using a while loop
a=("apple","banana","mango")
i=0
while i<len(a):
    print(a[i])
    i=i+1
```
