```python
# creating a function
def function():
    print("Hello world")
function()
```

```python
# Arguments    no. of arguments should be equal to called functions
def function(fname):
    print(fname + " hii")
    
function("Sam")
function("max")
function("henry")
```

```python
# Arbitary Arguments *args-used when we dont know no. of arguments
def function(*kids):
    print("The youngest child is " + kids[1])
function("ele","tom","Lisa")
```

```python
# Arbitrary keyword Arguments **kwargs
def function(**kid):
    print("His last name is" + kid["lname"])
function(fname = "sam", lname=" menekshaw")
```

```python
# list in arguments
def function(food):
    for x in food:
        print(x)
        
fruits = ["apple","banana","cherry"]
function(fruits)
```
