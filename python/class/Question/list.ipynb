```python
# write a program to accept input from user and print list elements and find minimum from list.

list1=[]
n=int(input("Enter any number: "))
for i in range(1,n+1):
    a=int(input("Enter elemets. "))
    list1.append(a)
print(list1)
print("min value: ",min(list1))
```
