```python
list1=[1,2,3,4,5]
def squares(x):
    return x**2
print(list(map(squares,list1)))
print(list(map(lambda x:x**2,list1)))
print([x**2 for x  in list1])
```

```python
# filter function
a=[1,2,3,5,7,9]
b=[2,3,6,7,9,8]
print(list(filter(lambda x:x in a,b)))
print([x for x in a if x not in b])
```
