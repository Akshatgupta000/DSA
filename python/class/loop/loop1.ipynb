```python
# print sum and average of a number

n=int(input("Enter no. "))
sum=0
i=1
while(i!=n+1):
    sum+=i
    i=i+1
print(sum)

average=sum/n
print(average)
```

```python
# print fibonacci series
t=int(input("How many terms? "))
n1,n2=0,1
count=0
print("Fibonacci sequence: ")
while(count < t):
    print(n1)
    nth=n1+n2
    n1=n2
    n2=nth
    count=count+1
```

```python
# program to find sum of digits of a number

n=int(input("Enter any no. "))
sum=0
while(n!=0):
    digit=int(n%10)
    sum=sum+digit
    n=n/10
print(sum)
```

```python
# program to check whether a number is palindrome or not

n=int(input("enter any number: "))
reverse=0
original=n
while(n>0):
    digit=n%10
    reverse=reverse*10 + digit
    n=n//10
    
if(original == reverse):
    print("palindrome")
else:
    print("not a palindrome")
print(reverse)
```
